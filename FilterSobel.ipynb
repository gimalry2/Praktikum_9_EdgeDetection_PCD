```python
import cv2  # Mengimpor pustaka OpenCV
import numpy as np  # Mengimpor pustaka NumPy
from matplotlib import pyplot as plt  # Mengimpor pustaka pyplot dari matplotlib
from skimage import data  # Mengimpor pustaka data skimage
```

```python
img = data.camera() # Memuat gambar dari skimage
```

```python
img_sobelx = cv2.Sobel(img, cv2.CV_8U, 1, 0, ksize=5)  # Menghitung gradien Sobel horizontal
img_sobely = cv2.Sobel(img, cv2.CV_8U, 0, 1, ksize=5)  # Menghitung gradien Sobel vertikal
img_sobel = img_sobelx + img_sobely  # Menggabungkan hasil gradien Sobel
```

```python
fig, axes = plt.subplots(4, 2, figsize=(20, 20))  # Membuat subplot 
ax = axes.ravel()  # Meratakan array 

ax[0].imshow(img, cmap='gray')  # Menampilkan citra input
ax[0].set_title("Citra Input")
ax[1].hist(img.ravel(), bins=256)  # Menampilkan histogram citra input
ax[1].set_title("Histogram Citra Input")

ax[2].imshow(img_sobelx, cmap='gray')  # Menampilkan citra output Sobel horizontal
ax[2].set_title("Citra Output")
ax[3].hist(img_sobelx.ravel(), bins=256)  # Menampilkan histogram citra output Sobel horizontal
ax[3].set_title("Histogram Citra Output")

ax[4].imshow(img_sobely, cmap='gray')  # Menampilkan citra output Sobel vertikal
ax[4].set_title("Citra Output")
ax[5].hist(img_sobely.ravel(), bins=256)  # Menampilkan histogram citra output Sobel vertikal
ax[5].set_title("Histogram Citra Output")

ax[6].imshow(img_sobel, cmap='gray')  # Menampilkan citra output Sobel gabungan
ax[6].set_title("Citra Output")
ax[7].hist(img_sobel.ravel(), bins=256)  # Menampilkan histogram citra output Sobel gabungan
ax[7].set_title("Histogram Citra Output")

fig.tight_layout()  # Mengatur tata letak subplot
```
