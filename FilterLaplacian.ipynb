```python
# Mengimpor pustaka
import cv2
from matplotlib import pyplot as plt

img0 = cv2.imread('orange.jpg',) # Membaca gambar
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY) # Konversi ke gray
img = cv2.GaussianBlur(gray, (3,3),0) # Menghilangkan noise
laplacian = cv2.Laplacian(img, cv2.CV_64F) # Konvolusi dengan kernel

# Menampilan dengan matplotlib
plt.subplot(1,2,1), plt.imshow(img, cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(laplacian, cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2 # Mengimpor pustaka

img = cv2.imread('orange.jpg',0) # Membaca citra
blur = cv2.GaussianBlur(img, (3,3),0) # Mengaplikasikan gausian
# Mengaplikasikan laplacian
laplacian = cv2.Laplacian(blur,cv2.CV_64F)
laplacian1 = laplacian/laplacian.max()

cv2.imshow('laplacian-gaussian',laplacian1) # Menampilkan citra filter Laplacian of Gaussian
cv2.waitKey(0) # Menunggu menekan tombol
```
