```python
import cv2  # Mengimpor pustaka OpenCV
import numpy as np  # Mengimpor pustaka NumPy
from matplotlib import pyplot as plt  # Mengimpor pustaka pyplot dari matplotlib
from skimage import data  # Mengimpor pustaka data skimage
```

```python
img = data.cat() # Memuat gambar dari skimage
```

```python
# prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])  # Kernel deteksi tepi secara horizontal
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])  # Kernel deteksi tepi secara vertikal

img_prewittx = cv2.filter2D(img, -1, kernelx)  # Mengaplikasikan Prewitt horizontal pada citra
img_prewitty = cv2.filter2D(img, -1, kernely)  # Mengaplikasikan Prewitt vertikal pada citra
img_prewitt = img_prewittx + img_prewitty  # Menggabungkan hasil  horizontal dan vertikal
```

```python
fig, axes = plt.subplots(4, 2, figsize=(20, 20))  # Membuat subplot 
ax = axes.ravel()  # Meratakan array 

# Citra Input
ax[0].imshow(img, cmap='gray')
ax[0].set_title("Citra Input")
# Histogram Citra Input
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title("Histogram Citra Input")

# Citra Output Prewitt X
ax[2].imshow(img_prewittx, cmap='gray')
ax[2].set_title("Citra Output Prewitt X")
# Histogram Citra Output Prewitt X
ax[3].hist(img_prewittx.ravel(), bins=256)
ax[3].set_title("Histogram Citra Output Prewitt X")

# Citra Output Prewitt Y
ax[4].imshow(img_prewitty, cmap='gray')
ax[4].set_title("Citra Output Prewitt Y")
# Histogram Citra Output Prewitt Y
ax[5].hist(img_prewitty.ravel(), bins=256)
ax[5].set_title("Histogram Citra Output Prewitt Y")

# Citra Output Prewitt
ax[6].imshow(img_prewitt, cmap='gray')
ax[6].set_title("Citra Output Prewitt")
# Histogram Citra Output Prewitt
ax[7].hist(img_prewitt.ravel(), bins=256)
ax[7].set_title("Histogram Citra Output Prewitt")

fig.tight_layout() # Mengatur tata letak subplot
```
