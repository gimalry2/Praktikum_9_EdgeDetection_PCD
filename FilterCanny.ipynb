```python
import cv2  # Mengimpor pustaka OpenCV
import numpy as np  # Mengimpor pustaka NumPy
from matplotlib import pyplot as plt  # Mengimpor pustaka pyplot dari matplotlib
from skimage import data  # Mengimpor pustaka data skimage
```

```python
img = data.astronaut() # Memuat gambar dari skimage
```

```python
#canny
img_canny = cv2.Canny(img,100,200) # Menerapkan deteksi tepi canny
```

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 20))  # Membuat subplot 
ax = axes.ravel()  # Meratakan array 

# Menampilkan citra input
ax[0].imshow(img, cmap='gray')
ax[0].set_title("Citra Input")

# Menampilkan histogram citra input
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title("Histogram Citra Input")

# Menampilkan citra output hasil deteksi tepi Canny
ax[2].imshow(img_canny, cmap='gray')
ax[2].set_title("Citra Output")

# Menampilkan histogram citra output
ax[3].hist(img_canny.ravel(), bins=256)
ax[3].set_title("Histogram Citra Output")
```
